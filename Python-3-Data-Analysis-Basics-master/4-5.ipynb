```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from ipywidgets import interact
```

```python
def move(n):
    print(" "*n + "oooo")
```

```python
move(10)
```

```python
interact(move, n=(1,100))
```

```python
x = np.linspace(0, 2*np.pi, 1000)
```

```python
def draw(n):
    y = np.sin(n*x)
    plt.plot(x, y, lw=3)
    plt.show()
```

```python
draw(10)
```

```python
interact(draw, n=(1., 50.))
```
