```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
x = np.linspace(0, 10, 50)
y = np.array(np.random.uniform(0, 10, 50))
x = x.reshape(50, 1)

from sklearn.linear_model import LinearRegression
regr = LinearRegression()
regr.fit(x, y)

plt.scatter(x, y)
plt.plot(x, regr.predict(x), 'r')
plt.show()
```
