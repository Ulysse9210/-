```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[-3, 2], [-6, 5], [3, -4], [2, -8]])
y = np.array([1, 1, 2, 2])
```

```python
plt.scatter([-3, -6, 3, 2], [2, 5, -4, -8], c=y)
```

```python
x
```

```python
x[2, 1]
```

```python
x[:,0]
```

```python
plt.scatter(x[:, 0], x[:, 1], c=y)
```

```python
plt.scatter(x[:, 0], x[:, 1], c=y, s=500)
```

```python
#SVM 支持向量機
from sklearn.svm import SVC
```

```python
clf = SVC()
```

```python
clf.fit(x, y)
```

```python
clf.predict([[-3,2]])
```

```python
clf.predict(x)
```

```python
clf.predict([[2.5, 10]])
```

```python
X, Y = np.meshgrid(np.linspace(-6, 3, 30), np.linspace(-8, 5, 30))
```

```python
plt.scatter(X, Y)
```

```python
#ravel拉平法
X = X.ravel()
Y = Y.ravel()
```

```python
plt.scatter(X, Y)
```

```python
#zip高級組合法
xx = [1, 2, 3, 4]
yy = [5, 6, 7, 8]
```

```python
list(zip(xx, yy))
```

```python
Z = clf.predict(list(zip(X, Y)))
```

```python
Z
```

```python
plt.scatter(X, Y, c=Z, s=50)
```
